##【1】比較演算子と論理演算子の復習
###比較演算子

>|意味|記述|
|:--:|:--:|
|より大きい|>|
|以上|>=|
|未満|<|
|以下|<=|
|等しい|==|
|等しくない|!=|

###論理演算子

>|意味|記述|値|
|:--:|:--:||
|否定(not)|!|論理値（True・False）の反転|
|かつ(and)|&&|両方がTrueであればTrue|
|または(or)|\|\||片方がTrueであればTrue|
|排他的論理和|xor|どちらか片方がTrueでもう片方がFalseであればTrueとなり、<br>両方ともTrueあるいは両方ともFalseの時はFalse|
|かつ(and)|&|演算結果をベクトルとして返す|
|または(or)|\||演算結果をベクトルとして返す|

>&、\| に関しては下記参照<br>

##【3】if～else文

###【実習4】入力された自然数が奇数ならば3倍して1を加えた数を、偶数ならば2で割った商を出力する。

````python
example7_04 <- function(n){
  if(n %% 2 == 0){ # nを2で割った余りが0のとき
    x <- n / 2 # %/%でもよい
  }else{
    x <- 3 * n + 1
  }
  return(x)
}
````

````python
example7_04(3)  #カッコの中の値を変えてみよう
````

##【4】if～else文の組み合わせ（else if文）
```R
if(条件1){
  条件1が成り立ったときの処理(複数行可) 
}else if(条件2){
  条件2が成り立ったときの処理(複数行可)
}else{
  すべての条件が成り立たなかったときの処理(複数行可)
}
```

````python
# 映画館の料金
x = 60
if(x < 6){
  cat("無料:\\0") 
}else if(x < 22){
  cat("学割:\\1,000")
}else if(x < 60){
  cat("普通料金:\\1,800")
}else{
  cat("シニア割:\\1,100")
}
````

###【実習5】入力された数値が正数、負数、零であることを判定し出力する

````python
example7_05 <- function(x){
  if(x > 0){
    cat("positive number\n")
  }else if(x < 0){
    cat("negative number\n")
  }else{
    cat("zero\n")
  }
  return(x)
}
````

````python
example7_05(0)  #カッコの中の値を変えてみよう
````

###【実習6】入力された自然数が2の倍数、3の倍数、5の倍数であることを判定し出力する。

````python
example7_05 <- function(n){
  if(n %% 2 == 0){ # nを2で割った余りが0のとき
    cat(n,"は2の倍数\n")
  }else if(n %% 3 == 0){ # nを3で割った余りが0のとき
    cat(n,"は3の倍数\n")
  }else if(n %% 5 == 0){ # nを5で割った余りが0のとき
    cat(n,"は5の倍数\n")
  }else{
    cat(n,"該当なし\n")
  }
  return(n)
}
````

````python
example7_05(4)  #カッコの中の値を変えてみよう
````

##再起呼び出し

###【実習8】自然数を入力し、1からその自然数までの平方和を出力する
image.png

````python
example7_08 <- function(n){
  if(n == 1){
    x <- 1
  }else{
    x <- n^2 + example7_08(n - 1) #自分自身の関数を呼び出すことを"再起呼び出し"という
  }
  return(x)
}
````

````python
example7_08(10)  #カッコの中の値を変えてみよう
````

##再起呼び出しとは
ある関数(example7_08)の定義の中に(example7_08)自身を呼び出している箇所があるとき、その呼び出しを再帰呼出し(recursive call)と言う。また、そのような定義は再帰的(recursive)である、と言う。
