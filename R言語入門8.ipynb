##【1】for文の入れ子構造
全章における反復処理の応用で、反復処理の中に反復処理を行うものを扱う。ここではfor文の中にfor文を使う処理を考える。

<font color = 'red'>for (変数1 in ベクトルやリスト){

　繰り返す処理（複数行可）</font>

　<font color = 'blue'>for (変数2 in ベクトルやリスト){

　　繰り返す処理（複数行可）</font>

　<font color = 'blue'>}</font>

<font color = 'red'>}</font>



```python
# for文の入れ子構造の例(掛け算九九)
for (x in 1:9){
  cat(x,'の段\n')
  for (y in 1:9){
    z <- x * y
    cat(x,'×',y,'＝',z,'\n')
  }
}
```

###【実習1】ベクトルの要素を昇順（小さい順）に並べ替える。このときのバブルソートで並べ替えるとする。

バブルソートとは、並び替え（ソート）のアルゴリズムのひとつで、隣り合うデータを順番に1つずつ大きさを比較・序列させる方式のことである。

image.png

参考文献 https://www.codereading.com/algo_and_ds/algo/bubble_sort.html

```python
# バブルソート
# リストの先頭から値を比較していき、昇順に並べ替えます。一番内側のループが終了すれば最大値がリストの終端に移動するので、次のループではリストの要素数をひとつ減らして比較していきます。
example9_01 <- function(vec){
  n <- length(vec)
  for (i in 1:(n - 1)){
    for (j in 1:(n - i)){
      print(vec)
      cat('    ',vec[j],'と',vec[j + 1],'を比較するよ\n')
      if(vec[j] > vec[j + 1]){
        cat('    ',vec[j],'は',vec[j + 1],'より大きいから場所を交換するよ\n')
        tmp <- vec[j + 1]
        vec[j + 1] <- vec[j]
        vec[j] <- tmp
      }
    }
  }
  return(cat('ソートが完了しました',vec))
}
```

```python
example9_01(c(1,7,3,5))
```

```python
# バブルソートを棒グラフでみてみよう
example9_02 <- function(vec){
  n <- length(vec)
  for (i in 1:(n - 1)){
    for (j in 1:(n - i)){
      names(vec) <- vec 
      vec.col <- rep('white',times = n)
      vec.col[j] <- 'yellow'
      if(vec[j] > vec[j + 1]){
        vec.col[j + 1] <- 'red'
        barplot(vec,col=vec.col)
        cat('【処理',i,'の',j,'】',vec[j],'と',vec[j + 1],'を比較するよ\n')
        cat('    ',vec[j],'は',vec[j + 1],'より大きいから場所を交換するよ\n')
        tmp <- vec[j + 1]
        vec[j + 1] <- vec[j]
        vec[j] <- tmp
      }else{
        vec.col[j+1] <- 'blue'
        barplot(vec,col=vec.col)
        cat('【処理',i,'の',j,'】',vec[j],'と',vec[j+1],'を比較するよ\n')
        cat('      そのままだね\n')
      }
    }
  }
  return(cat('ソートが完了しました',vec))
}
```

```python
example9_02(c(1,7,3,5))
```
