##【1】for文の入れ子構造（ネスト）
全章における反復処理の応用で、反復処理の中に反復処理を行うものを扱う。ここではfor文の中にfor文を使う処理を考える。

<font color = 'red'>for (変数1 in ベクトルやリスト){

　繰り返す処理（複数行可）</font>

　<font color = 'blue'>for (変数2 in ベクトルやリスト){

　　繰り返す処理（複数行可）</font>

　<font color = 'blue'>}</font>

<font color = 'red'>}</font>



```python
# for文の入れ子構造の例(掛け算九九)
for (x in 1:9){
  cat(x,'の段\n')
  for (y in 1:9){
    z <- x * y
    cat(x,'×',y,'＝',z,'\n')
  }
}
```

###【実習1】ベクトルの要素を昇順（小さい順）に並べ替える。このときのバブルソートで並べ替えるとする。

バブルソートとは、並び替え（ソート）のアルゴリズムのひとつで、隣り合うデータを順番に1つずつ大きさを比較・序列させる方式のことである。

```python
# バブルソート
# リストの先頭から値を比較していき、昇順に並べ替えます。一番内側のループが終了すれば最大値がリストの終端に移動するので、次のループではリストの要素数をひとつ減らして比較していきます。
example9_01 <- function(vec){
  n <- length(vec)
  for (i in 1:(n - 1)){
    for (j in 1:(n - i)){
      print(vec)
      cat('    ',vec[j],'と',vec[j + 1],'を比較\n')
      if(vec[j] > vec[j + 1]){
        cat('    ',vec[j],'は',vec[j + 1],'より大きいから',vec[j],'と',vec[j + 1],'を交換するよ\n')
        replacevalue1 <- vec[j]
        replacevalue2 <- vec[j + 1]
        vec[j] <- replacevalue2
        vec[j + 1] <- replacevalue1
      }
    }
  }
  return(cat('ソートが完了しました',vec))
}
```

```python
example9_01(c(1,7,3,5))
```

##バブルソートを視覚的にみる

```python
# バブルソートを棒グラフでみてみよう
example9_02 <- function(vec){
  n <- length(vec)
  for (i in 1:(n - 1)){
    for (j in 1:(n - i)){
      names(vec) <- vec 
      vec.col <- rep('white',times = n)
      vec.col[j] <- 'yellow'
      if(vec[j] > vec[j + 1]){
        vec.col[j + 1] <- 'red'
        barplot(vec,col=vec.col)
        cat('【処理',i,'の',j,'】',vec[j],'と',vec[j + 1],'を比較\n')
        cat('    ',vec[j],'は',vec[j + 1],'より大きいから',vec[j],'と',vec[j + 1],'を交換するよ\n')
        replacevalue1 <- vec[j]
        replacevalue2 <- vec[j + 1]
        vec[j] <- replacevalue2
        vec[j + 1] <- replacevalue1
      }else{
        vec.col[j+1] <- 'blue'
        barplot(vec,col=vec.col)
        cat('【処理',i,'の',j,'】',vec[j],'と',vec[j+1],'を比較\n')
        cat('      そのまま\n')
      }
    }
  }
  return(cat('ソートが完了しました',vec))
}
```

```python
example9_02(c(1,7,3,5))
```

###【実習2】ベクトルの要素を昇順（小さい順）に並べ替える。このときの選択ソートで並べ替えるとする。

>選択ソートとは、<font color ='red'>対象となるデータの中</font>から最小値（または最大値）を探して、先頭の値と交換する作業を繰り返すことで全体を整列させていく手法です。

image.png

Copyright 2007-2021 codereading.com

参考文献 https://www.codereading.com/algo_and_ds/algo/selection_sort.html



```python
# 選択ソート
example9_03 <- function(vec){
  n <- length(vec)
  for (i in 1:(n - 1)){
    tmp <- i
    print(vec)
    for (j in (i + 1):n){
      cat('    ',vec[tmp],'と',vec[j],'を比較するよ\n')
      if(vec[tmp] > vec[j]){
        cat('     ',vec[tmp],'は',vec[j],'より大きいから',vec[j],'を先頭(',vec[tmp],')と交換\n')
        tmp <- j
      }
    }
    replacevalue1 <- vec[i]
    replacevalue2 <- vec[tmp]
    vec[i] <- replacevalue2
    vec[tmp] <- replacevalue1
  }
  return(cat('ソートが完了しました',vec))
}
```

```python
example9_03(c(3,9,6,1,2))
```
